```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Plot global notation
cmaps = ['tab10',['Encontradas','Sobrantes','Coincidencias','No encontradas']]

gradient = np.ones((1,1))

nrows = len(cmaps[1])
figh = 2
fig, axes = plt.subplots(nrows=nrows, figsize=(1, figh))
fig.subplots_adjust(top=figh/3, bottom=.15/figh, left=0.5, right=0.99)


for ax, name, i in zip(axes, cmaps[1], range(5)):   
    ax.imshow(gradient*i, aspect='auto', cmap=cmaps[0],vmin=0,vmax=9)
    ax.text(-.2, .5, name, va='center', ha='right', fontsize=10, transform=ax.transAxes)

for ax in axes:
    ax.set_axis_off()

plt.show()
# plot donut chart

size = 0.3
cmap = plt.get_cmap("tab10")
outer_colors = cmap(np.array([3,0]))
inner_colors = cmap(np.array([3,2,1]))


data = np.array([
    [[26,195],[26,112,83]],
    [[10,140],[10,73,67]],
    [[12,202],[12,81,121]],
    [[5,403],[5,132,271]],
    [[8,348],[8,91,257]]])


for i in range(data.shape[0]):
    fig, ax = plt.subplots()

    sistem = data[i][0]
    expert = data[i][1]

    ax.pie(sistem, radius=1,colors=outer_colors,wedgeprops=dict(width=size, edgecolor='w'))
    ax.pie(expert, radius=1-size, colors=inner_colors,wedgeprops=dict(width=size, edgecolor='w'))

    ax.set(aspect="equal")
    plt.show()
```
